```python
#@title Código para gerar o relatório
#importanto a biblioteca que lê e cria docs:
from docxtpl import DocxTemplate 
#importando a biblioteca que lê a planilha:
import pandas as pd 

#atribuindo a serie
serie = "9th" #@param ["9th","8th"]

#atribuindo os caminhos dos arquivos:
caminhotabela = f"/content/drive/MyDrive/ReportGen/{serie}/Relatorio_PrimeiroTrimestre_{serie}.xlsx"
pasta_para_salvar = f'/content/drive/MyDrive/ReportGen/{serie}/Gerados'
caminhotemplate = f"/content/drive/MyDrive/ReportGen/{serie}/Template_Relatorio_{serie}.docx"

tabela = pd.read_excel(caminhotabela,sheet_name=0)

for linha in tabela.index:
    #pegando o código, nome e comentário do aluno:
    tabela = pd.read_excel(caminhotabela,sheet_name=0) 
    #lendo o template do documento
    documento = DocxTemplate(caminhotemplate)
    #procurando a coluna e atribuindo o valor da linha para cada variável:
    codigo = tabela.loc[linha, "code"]
    nome = tabela.loc[linha, "Name"]
    math = tabela.loc[linha, "Math"]
    science = tabela.loc[linha, "Science"]
    globalperspective = tabela.loc[linha, "Global"]
    robotic = tabela.loc[linha, "Robotics"]
    gastronomy = tabela.loc[linha, "Gastronomia"]
    generalactivity = tabela.loc[linha, "AtividadesGerais"]

    #biblioteca correlacionando as tags que estão no doc com as variáveis:
    referencias = {
        "NNNN": nome,
        "MMMM": math,
        "SSSS": science,
        "GGGG": globalperspective,
        "RRRR": robotic,
        "GAGA": gastronomy,
        "AAAA": generalactivity,
        }
    #abrindo a segunda aba com os unit tests:
    tabela = pd.read_excel(caminhotabela,sheet_name=1) #lendo a planilha e guardando na variável tabela
    #atribuindo o valor de cada unit test:
    UnitMath1 = tabela.loc[linha, "UnitMath1"]
    UnitMath2 = tabela.loc[linha, "UnitMath2"]
    UnitMath3 = tabela.loc[linha, "UnitMath3"]
    UnitMath4 = tabela.loc[linha, "UnitMath4"]
    UnitMath5 = tabela.loc[linha, "UnitMath5"]
    UnitScience1 = tabela.loc[linha, "UnitScience1"]
    UnitScience2 = tabela.loc[linha, "UnitScience2"]
    UnitScience3 = tabela.loc[linha, "UnitScience3"]
    UnitScience4 = tabela.loc[linha, "UnitScience4"]
    UnitScience5 = tabela.loc[linha, "UnitScience5"]
    UnitGlobal1 = tabela.loc[linha, "UnitGlobal1"]
    UnitGlobal2 = tabela.loc[linha, "UnitGlobal2"]
    UnitGlobal3 = tabela.loc[linha, "UnitGlobal3"]
    UnitGlobal4 = tabela.loc[linha, "UnitGlobal4"]
    UnitGlobal5 = tabela.loc[linha, "UnitGlobal5"]

    #adicionando ao dicionário:
    referencias = {
        "NNNN": nome,
        "MMMM": math,
        "SSSS": science,
        "GGGG": globalperspective,
        "RRRR": robotic,
        "GAGA": gastronomy,
        "AAAA": generalactivity,
        #adicionando os valores dos units da segunda aba:
        "UUM1": UnitMath1,
        "UUM2": UnitMath2,
        "UUM3": UnitMath3,
        "UUM4": UnitMath4,
        "UUM5": UnitMath5,
        "UUS1": UnitScience1,
        "UUS2": UnitScience2,
        "UUS3": UnitScience3,
        "UUS4": UnitScience4,
        "UUS5": UnitScience5,
        "UUG1": UnitGlobal1,
        "UUG2": UnitGlobal2,
        "UUG3": UnitGlobal3,
        "UUG4": UnitGlobal4,
        "UUG5": UnitGlobal5,
    }
    #abrindo a terceira aba com os progress check:
    tabela = pd.read_excel(caminhotabela,sheet_name=2) #lendo a planilha e guardando na variável tabela
    #atribuindo o valor de cada progress check:
    progressmath = tabela.loc[linha, "ProgressScoreMath"] 
    mathgradefinal = tabela.loc[linha, "MathGradeFinal"]
    colormath = tabela.loc[linha, "ColorMath"]
    feedmath =  tabela.loc[linha, "FeedbackMath"]

    progressscience = tabela.loc[linha, "ProgressScoreScience"] 
    sciencegradefinal = tabela.loc[linha, "ScienceGradeFinal"]
    colorscience = tabela.loc[linha, "ColorScience"]
    feedscience =  tabela.loc[linha, "FeedbackScience"]

    progressglobal = tabela.loc[linha, "ProgressScoreGlobal"] 
    globalgradefinal = tabela.loc[linha, "GlobalGradeFinal"]
    colorglobal = tabela.loc[linha, "ColorGlobal"]
    feedglobal =  tabela.loc[linha, "FeedbackGlobal"]

    #adicionando ao dicionário:
    referencias = {
        "NNNN": nome,
        "MMMM": math,
        "SSSS": science,
        "GGGG": globalperspective,
        "RRRR": robotic,
        "GAGA": gastronomy,
        "AAAA": generalactivity,
        #math
        "UUM1": UnitMath1,
        "UUM2": UnitMath2,
        "UUM3": UnitMath3,
        "UUM4": UnitMath4,
        "UUM5": UnitMath5,
        #science
        "UUS1": UnitScience1,
        "UUS2": UnitScience2,
        "UUS3": UnitScience3,
        "UUS4": UnitScience4,
        "UUS5": UnitScience5,
        #global
        "UUG1": UnitGlobal1,
        "UUG2": UnitGlobal2,
        "UUG3": UnitGlobal3,
        "UUG4": UnitGlobal4,
        "UUG5": UnitGlobal5,
        #adicionando os valores do progression, valor final, cor e feed de cada área:
        #math
        "PPPM": progressmath,
        "OVRSCRMATH": mathgradefinal,
        "COLORMATH": colormath,
        "FEEDMATH" : feedmath,
        #science
        "PPPS": progressscience,
        "OVRSCRSCIENCE": sciencegradefinal,
        "COLORSCIENCE": colorscience,
        "FEEDSCIENCE" : feedscience,
        #global
        "PPPG": progressglobal,
        "OVRSCRGLOBAL": globalgradefinal,
        "COLORGLOBAL": colorglobal,
        "FEEDGLOBAL" : feedglobal,

    }
    #Substituindo cada tag pelo valor da variável
    documento.render(referencias)
    #salvando em um documento com o nome para cada linha
    documento.save(f'{pasta_para_salvar}/{codigo}.docx')
```
