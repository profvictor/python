```python
#@title Gerador de Lesson Plan
#importanto a biblioteca que lê e cria docs:
from docxtpl import DocxTemplate
#importando a biblioteca que lê a planilha e dando um nickname de pd para o pandas:
import pandas as pd
#definindo para ler a primeira aba da planilha (0 é a primeira aba, 1 a segunda...):
sheet = 0
#lendo a planilha e guardando na variável tabela:
tabela = pd.read_excel("/content/drive/MyDrive/LessonPlanGen/LessonPlanInput.xlsx",sheet_name=sheet)

for linha in tabela.index:
    #lendo o template do documento
    documento = DocxTemplate("/content/drive/MyDrive/LessonPlanGen/TemplateLessonplan.docx")
    #procurando a coluna e atribuindo o valor da linha para cada variável:
    day = tabela.loc[linha, "day"]
    framework = tabela.loc[linha, "framework"]
    teacherguide = tabela.loc[linha, "teacherguide"]
    vocabulary = tabela.loc[linha, "vocabulary"]
    learnersbook = tabela.loc[linha, "learnersbook"]
    question = tabela.loc[linha, "question"]
    theory = tabela.loc[linha, "theory"]
    exercises = tabela.loc[linha, "exercises"]
    evidence = tabela.loc[linha, "evidence"]
    link = tabela.loc[linha, "link"]
    archive = tabela.loc[linha, "Lesson"]
    #criando um dicionário e correlacionando as tags que estão no doc com as variáveis:
    referencias = {
        "DDDD": day,
        "FFFF": framework,
        "GGGG": teacherguide,
        "VVVV": vocabulary,
        "LLLL": learnersbook,
        "QQQQ": question,
        "TTTT": theory,
        "EEEE": exercises,
        "ZZZZ": evidence,
        "WWWW": link,
        "YYYY": archive,
    }
    #Substituindo cada tag pelo valor da variável com a função render:
    documento.render(referencias)
    #salvando em um documento com o nome para cada linha usando a tag archive para nomear:
    documento.save(f'/content/drive/MyDrive/LessonPlanGen/8th/{archive}.docx')

#definindo para ler a segunda aba da planilha
sheet = 1
tabela = pd.read_excel("/content/drive/MyDrive/LessonPlanGen/LessonPlanInput.xlsx",sheet_name=sheet)

#Idem do procedimento anterior só que para o 9th:
for linha in tabela.index:
    documento = DocxTemplate("/content/drive/MyDrive/LessonPlanGen/TemplateLessonplan.docx")
    day = tabela.loc[linha, "day"]
    framework = tabela.loc[linha, "framework"]
    teacherguide = tabela.loc[linha, "teacherguide"]
    vocabulary = tabela.loc[linha, "vocabulary"]
    learnersbook = tabela.loc[linha, "learnersbook"]
    question = tabela.loc[linha, "question"]
    theory = tabela.loc[linha, "theory"]
    exercises = tabela.loc[linha, "exercises"]
    evidence = tabela.loc[linha, "evidence"]
    link = tabela.loc[linha, "link"]
    archive = tabela.loc[linha, "Lesson"]
    referencias = {
        "DDDD": day,
        "FFFF": framework,
        "GGGG": teacherguide,
        "VVVV": vocabulary,
        "LLLL": learnersbook,
        "QQQQ": question,
        "TTTT": theory,
        "EEEE": exercises,
        "ZZZZ": evidence,
        "WWWW": link,
        "YYYY": archive,
    }
    documento.render(referencias)
    documento.save(f'/content/drive/MyDrive/LessonPlanGen/9th/{archive}.docx')

    #definindo para ler a terceira aba da planilha
sheet = 2
tabela = pd.read_excel("/content/drive/MyDrive/LessonPlanGen/LessonPlanInput.xlsx",sheet_name=sheet)

#Idem do procedimento anterior só que para o 9th:
for linha in tabela.index:
    documento = DocxTemplate("/content/drive/MyDrive/LessonPlanGen/TemplateLessonplan.docx")
    day = tabela.loc[linha, "day"]
    framework = tabela.loc[linha, "framework"]
    teacherguide = tabela.loc[linha, "teacherguide"]
    vocabulary = tabela.loc[linha, "vocabulary"]
    learnersbook = tabela.loc[linha, "learnersbook"]
    question = tabela.loc[linha, "question"]
    theory = tabela.loc[linha, "theory"]
    exercises = tabela.loc[linha, "exercises"]
    evidence = tabela.loc[linha, "evidence"]
    link = tabela.loc[linha, "link"]
    archive = tabela.loc[linha, "Lesson"]
    referencias = {
        "DDDD": day,
        "FFFF": framework,
        "GGGG": teacherguide,
        "VVVV": vocabulary,
        "LLLL": learnersbook,
        "QQQQ": question,
        "TTTT": theory,
        "EEEE": exercises,
        "ZZZZ": evidence,
        "WWWW": link,
        "YYYY": archive,
    }
    documento.render(referencias)
    documento.save(f'/content/drive/MyDrive/LessonPlanGen/7th/{archive}.docx')
```
